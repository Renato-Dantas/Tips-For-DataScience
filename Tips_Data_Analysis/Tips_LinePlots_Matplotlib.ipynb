```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
y = [2,4,6,8,10,12,24,48]
x = [2,20,25,30,35,40,45,50]
```

```python
plt.plot(x,y)
```

## Putting a title, axis labels and some details on plot

```python
# Defines a title for the plot 
plt.title('Line Plot',fontdict={'fontsize':15}, loc='center', pad=10)
# plt.title(title, dict of options {see documentation}, loc = position, pad = distance of the edge)

plt.xlabel('X axis', fontsize = 15)
plt.ylabel('Y axis', fontsize = 20)

plt.plot(x,y,'g--*')
# We can set the line color, line way and markers with a simple description like 'g--*' (green - dashed line - marker)
```

## Creating a Subplot

**The subplot allow us to make two or more plots in the same image**

```python
plt.subplot(1,2,1)
# Here is defined the n° of rows, n° of columns, and what plot will be edited - ex (1,2,1): (1 row, 2 columns, first plot)

plt.plot(x,y,'r--o')

# Now we set a subplot for the second plot (1,2,2)
plt.subplot(1,2,2)
plt.plot(y,x, 'b--*')

```

### We also can put the lines in the same plot

```python
figure, axis = plt.subplots()

axis.plot(y,x, 'r-o', label = 'x for y') # Setting a label to the line
axis.plot(x,y,'g--x', label = 'y for x') # Setting a label to the line

axis.legend(loc = 'center') # Calling the labels inside the plot and choosing the label position
```

### Settings for two plots using subplots

```python
fig, ax = plt.subplots(1,2) #(n° rows, n° columns)

ax[0].plot(x,y, 'black')
ax[1].plot(y,x,'r--')

ax[0].set_title('First Plot')
ax[1].set_title('Second Plot')

ax[0].set_xlabel('X',fontsize = 15)
ax[1].set_xlabel('Second X',fontsize = 15)

ax[0].set_ylabel('Y Axis', fontsize = 20)
```
