# Merge, Join and Concatenate dataframes!

**Sometimes We'll need to join two dataframes, for it we can use some smarth methods!**

First, lets create some data

To start,**import the Pandas library**

```python
import pandas as pd
```

```python
A = pd.DataFrame({'A':['A0','A1','A2','A3'],
                'B':['B0','B1','B2','B3'],
                'C':['C0','C1','C2','C3']})

A.head()
```

```python
B = pd.DataFrame({'A':['A4','A5','A6','A7'],
                'B':['B4','B5','B6','B7'],
                'C':['C4','C5','C6','C7']})

B.head()
```

```python
C = pd.DataFrame({'A':['A8','A9','A10','A11'],
                'B':['B8','B9','B10','B11'],
                'C':['C8','C9','C10','C11']})
C.head()
```

## Cool! Let's do a Concat!

**A concat basicly glue one dataframe to another, see this example below**

```python
# Using .concat, we pass to this function a list with the datasets to 'glue'

# To save this change, use the parameter 'inplace = True'

# Remember: axis = 0 will make the 'glue' in rows, to make the glue in columns, use 'axis = 1'

pd.concat([A,B], axis = 0)
```

```python
pd.concat([A,C,B], axis = 1)
```

## Using the MERGE

**The merge function allows us to merge dataframes with a commom key**

This key must be the same in both dataframes

Let's create this "Key" column

```python
A['Key'] = [10,20,30,40]
B['Key'] = [10,20,30,40]
```

```python
A
```

```python
B
```

**With a common column, we can merge the dataframes using the same concept that SQL queries uses**

```python
# To this funcion, we pass the data1, data2, the method and where to merge
# pd.merge(data1,data2, how = method, on = column to merge)

pd.merge(A,B, how = 'inner', on= 'Key')
```

If we have more than one common column, on the parameter 'on' we can pass a list with the columns to merge

## Let's use the JOIN

**The join will just put a dataframe inside the other, if the columns without common values, it'll get a Nan value**

```python
to_join = pd.DataFrame({'D':[10,20,30,40], 'E':['A','B','C','D']})
to_join
```

```python
# Here we just put the new dataframe (to_join) in the left of our dataframe A

# There's more ways to make this join, to see all, you can press 'tab' inside the function to read the options (only on jupyter)

A.join(to_join)
```
